```python
#Ordenación de una lista 
def   bubble_sort_f ( lista ):
    n =  0
    while  n <  len (lista) -1 :
      n =  0
      for  i  in   range ( 0 , len (lista) -1 ):
        if  lista[i] > lista[i+ 1 ]:
          lista[i],lista[i+ 1 ]=lista[i+ 1 ],lista[i]
        else :
          n +=  1
  
    return  lista

print(bubble_sort_f([3,4,1,7]))
```

```python
# FORO - Algoritmo voraz: cambio con monedas [25,10,5,1]
def cambioMonedas(total, tipos_monedas, solucion, acum):
    for i in range(len(tipos_monedas)):
      if(acum < total):
        queda_actual = total - tipos_monedas[i]
        # print("total",total, "queda",queda_actual)
        if(queda_actual >= 0):
            # print("-> coje moneda de", tipos_monedas[i])
            solucion[i] += 1
            acum += tipos_monedas[i]
            solucion, acum = cambioMonedas(queda_actual, tipos_monedas, solucion, acum)  
    
    return solucion, acum

sol = cambioMonedas(300, [25,10,5,1], [0,0,0,0], 0)
print(sol)
```

```python
# Decorador que mide en ms el tiempo que tarda un método
import time

def timer_decorator(method):
    def measure_time(*args, **kw):
        time_begin = time.time()
        result = method(*args, **kw)
        time_end = time.time()
        ms_elapsed = (time_end - time_begin) * 1000

        if 'log_time' in kw:
            name = kw.get('log_name', method.__name__.upper())
            kw['log_time'][name] = ms_elapsed
        else:
            print(method.__name__, "lasted", ms_elapsed, "ms")

        return result
    return measure_time


# se declara el decorador sobre el método que se quiera medir
@timer_decorator 
def method_to_measure(**kwargs):
  time.sleep(1.16) # Prueba para que se pare un tiempo


# se llama al método con el diccionario donde se guardará la info
# si no se manda el diccionario se mostrará por pantalla
time_info ={}
method_to_measure(log_time=time_info)

print(time_info)
```

```python
# La siguiente funcioón recorre el vector de izquieda a derecha. 
# El primer bucle coge un elemento mientras que el segundo compara ese elemento con todos los demás. 
# El segundo bucle comienza desde la posicición del primer elemento encontrado hasta el final.

@timer_decorator 
def   cercanosFuerza ( lista ):
   # n =  0
    # Inicializamos las variables con los dos primeros números de la lista
    menor = abs(lista[0] - lista[1])
    numeros = (lista[0] , lista[1])
    
    for  n  in range(0 ,  len (lista)) :
        numero1 = lista[n]
        for  i  in   range ( n+1 , len (lista) ):
            aux = abs(lista[n] - lista[i])
            #print(aux)
            if aux < menor :
                menor = aux
                numero2 =  lista[i]
                # print (numeros)
       
    return  menor, [numero1, numero2]


print(cercanosFuerza(LISTA_1D))
```

## Divide y Venceras 

```python
# La siguiente función es la parte mas pequeña del problema 
# Recibe como parametro un vector de 1D, el limite inferior, el limite superior y el valor a comparar con cada uno

def divideyvenceras1D(vec, inf,sup, comp):
  
    if inf== sup :
        return abs(vec[sup-1] -comp) , vec[sup-1] 
    else:
        mitad= ((sup+inf)//2)
        #print(mitad)
        #print(str(inf) +'  '+ str(sup) )
        maxIzq, numero1 = divideyvenceras1D(vec, inf,mitad,comp)
        maxDcha,numero2 = divideyvenceras1D(vec, mitad+1 , sup,comp)
        
        if maxIzq < maxDcha: 
            return maxIzq,numero1
        else:
            return maxDcha,numero2
       
    
print(divideyvenceras1D([20,50,90 ,40, 66,70,9,80],7,8,66))
```

```python
# En esta función se resuelve el problema recorriendo la lista y comparando con cada elemento el resto del vector
# Se Utiliza el algoritmo de Divide y Venceras para resolver cada comparación. 
# Es necesario utilizar 
 
@timer_decorator 
def   cercanosDivide1D ( lista):
    # Inicializamos las variables con los dos primeros números de la lista
    menor = abs(lista[0] - lista[1])
    numeros = (lista[0] , lista[1])

    for  n  in range(0 ,  len (lista)-1) :
        numero2= lista[n]
        aux, numero1 = divideyvenceras1D(lista, n+2, len (lista),lista[n] )
        #print(n,aux, numero1, numero2)
        if aux < menor : 
            menor = aux
            numeros = [numero1,numero2]

    return  menor, numeros


print(cercanosDivide1D(LISTA_1D))
```
